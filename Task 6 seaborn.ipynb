# Seaborn -
Seaborn is a powerful and versatile data visualization library in Python, built on top of the popular Matplotlib library. It provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn is particularly useful for visualizing complex datasets and for creating aesthetically pleasing visualizations with minimal code. 

```python
# 1. Distribution Plot
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = sns.load_dataset('tips')
sns.histplot(df['total_bill'], kde=True)
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2.Box plot
sns.boxplot(x='day', y='total_bill', data=df)
plt.title('Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
# 3.violin plot
sns.violinplot(x='day', y='total_bill', data=df)
plt.title('Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
# 4.Bar plot
sns.barplot(x='day', y='total_bill', data=df, estimator=sum)
plt.title('Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
# 5.scatter plot
sns.scatterplot(x='total_bill', y='tip', data=df, hue='time')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# 6.line plot
sns.lineplot(x='size', y='total_bill', data=df)
plt.title('Total Bill by Size')
plt.xlabel('Size')
plt.ylabel('Total Bill')
plt.show()
```

```python
# 7.heatmap
import pandas as pd
tips_numeric=pd.get_dummies(df,drop_first=True)
c=tips_numeric.corr()
sns.heatmap(c,annot=True,cmap="rainbow")
plt.show()
```

```python
# 8.pair plot
sns.pairplot(df, hue='time')
plt.show()
```

```python
# 9.facet grid
g = sns.FacetGrid(df, col="time", row="day")
g.map(sns.histplot, "total_bill")
plt.show()
```

```python
# 10.regression plot
import seaborn as sns
import matplotlib.pyplot as plt
# Load a sample dataset
df = sns.load_dataset('tips')
# Create a regression plot
sns.regplot(x='total_bill', y='tip', data=df)
plt.title('Regression Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# 11.residual plot
sns.residplot(x='total_bill', y='tip', data=df, lowess=True)
plt.title('Residual Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Residuals')
plt.show()
```
